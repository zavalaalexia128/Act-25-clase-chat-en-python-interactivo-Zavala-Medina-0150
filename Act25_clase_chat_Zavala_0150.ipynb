#### Crear la clase "Chat", con 3 atributos (usuario1, usuario2, mensajes)
#### 2 funciones leerDatos(), mostrarDatos() y 2 objetos (chat1, chat2).
#### que sea interctivo --> input
#### no usar while true, if, for, lista y diccionario

## Chat
![](https://raw.githubusercontent.com/zavalaalexia128/Act-25-clase-chat-en-python-interactivo-Zavala-Medina-0150/refs/heads/main/InfoChat0150.png)

```python
class Chat:
    def __init__(self, usuario1, usuario2, mensajes):
        self.usuario1 = usuario1
        self.usuario2 = usuario2
        self.mensajes = mensajes

    def leerdatos(self):
        print("Configurando el chat entre " + self.usuario1 + " y " + self.usuario2)
        self.mensajes = input("Introduce el historial de mensajes para este chat: ")

    def mostrardatos(self):
        print("\n--- Datos del Chat ---")
        print("Usuario 1: " + self.usuario1)
        print("Usuario 2: " + self.usuario2)
        print("Ultimos mensajes: " + self.mensajes)
        print("----------------------")

print("Registro de Chat 1")
u1_chat1 = input("Introduce el nombre del primer usuario: ")
u2_chat1 = input("Introduce el nombre del segundo usuario: ")
chat1 = Chat(u1_chat1, u2_chat1, "")
chat1.leerdatos()

print("\nRegistro de Chat 2")
u1_chat2 = input("Introduce el nombre del primer usuario: ")
u2_chat2 = input("Introduce el nombre del segundo usuario: ")
chat2 = Chat(u1_chat2, u2_chat2, "")
chat2.leerdatos()

chat1.mostrardatos()
chat2.mostrardatos()

print("\nPrograma realizado por Zavala Medina Alexia Paola NL= 54")
```
